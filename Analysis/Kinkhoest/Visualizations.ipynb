# Visualizations

### We start with simple Visualizations for The Incidence Data

First we import the necessary libraries and then the csv file that includes the preprocessed data for the Brabant GGDs and Brabant Google Trend.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import datetime as dt
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
os.chdir('C:\\Users\\ekoulier\\Desktop\\GGD\\New Data\\Central_Folder')
alldata = pd.read_csv('br_regions_gt.csv')
os.chdir(cwd)
```

```python
del alldata['Unnamed: 0']
alldata['Date'] = pd.to_datetime(alldata['Date'], format='%Y-%m')
alldata.head()
```

```python
for i in ['HVB', 'WB', 'BZO', 'Trends']:
    plt.plot(alldata['Date'], alldata[i])
    #plt.title('Incidences in '+i)
    plt.xlabel('Year')
    plt.ylabel('Number of Incidences')
    plt.axvline(dt.datetime(2013,11,1), linestyle = 'dashed', color = 'black', linewidth = 0.8)
    plt.savefig('Incidences in '+i, dpi = 800)
    plt.show()
```

```python
plt.plot(alldata['Date'], alldata['Trends'])
plt.title('North Brabant Google Trends Data')
plt.xlabel('Year')
plt.ylabel('Number of Incidences')
plt.axvline(dt.datetime(2013,11,1), linestyle = 'dashed', color = 'black', linewidth = 0.8)
plt.savefig('Google Search', dpi = 1000)
plt.show()
```
